```python
sql_query = "SELECT * FROM cost JOIN outturn ON cost.country = outturn.country and cost.type = outturn.type and cost.year=outturn.year;"

df = pd.read_sql(sql_query, conn)
df
```

```python
df = df.T.drop_duplicates().T
```

```python
columns_to_drop = ['index']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# for further manipulation, combining year and type into one distinct column 
df['year_type'] = df['year'].astype(str) + df['type'].apply(lambda x: '_1' if x == 'summer' else '_0')
columns_to_drop = ['year', 'type']
df = df.rename(columns={'country': 'country_code'})
desired_column_order = ['year_type','games', 'country_code','events', 'cost_per_event_mio_usd','athletes', 'cost_per_athlete_mio_usd', 'cost_billion_usd']

df = df[desired_column_order]
df
```

```python
sql_query = "SELECT season, year, city, sport, discipline, country_code, event, COUNT(*) AS medal_count FROM medals GROUP BY season, year, city, sport, discipline, country_code, event;"
dfm = pd.read_sql(sql_query, conn)
conn.close()
dfm.head()
```

```python
dfmc = dfm.groupby(['year', 'season', 'country_code'])['medal_count'].sum().reset_index()
dfmc['year_type'] = dfmc['year'].astype(str) + dfmc['season'].apply(lambda x: '_1' if x == 'summer' else '_0')
columns_to_drop = ['year', 'season']
dfmc.drop(columns=columns_to_drop, inplace=True)

dfmc.head()
```

```python
dfmc[dfmc['country_code']=='NOR']
```

```python
# Step 1: Merge the DataFrames on 'year_type' and 'country_code'
host = pd.merge(df, dfmc, on=['year_type', 'country_code'], how='inner')

# Step 2: Create a function to calculate medal counts
def calculate_medals(row):
    year_type = row['year_type']
    if year_type == 1992_0: 
        next_year_type = f"{int(year_type[:4])+2}{year_type[4:]}"
    elif year_type == 1994_0: 
        prev_year_type = f"{int(year_type[:4])-2}{year_type[4:]}"  
    else:
        prev_year_type = f"{int(year_type[:4])-4}{year_type[4:]}"  # Get year_type 4 years before
        next_year_type = f"{int(year_type[:4])+4}{year_type[4:]}"  # Get year_type 4 years after

    # Calculate mbefore_hosting, mhosting, and mafter_hosting
    mbefore_hosting = dfmc[(dfmc['year_type'] == prev_year_type) & (dfmc['country_code'] == row['country_code'])]['medal_count'].values
    mhosting = dfmc[(dfmc['year_type'] == year_type) & (dfmc['country_code'] == row['country_code'])]['medal_count'].values
    mafter_hosting = dfmc[(dfmc['year_type'] == next_year_type) & (dfmc['country_code'] == row['country_code'])]['medal_count'].values

    return pd.Series({
        'mbefore_hosting': mbefore_hosting[0] if mbefore_hosting.size > 0 else np.nan,
        'mhosting': mhosting[0] if mhosting.size > 0 else np.nan,
        'mafter_hosting': mafter_hosting[0] if mafter_hosting.size > 0 else np.nan
    })

# Step 3: Apply the function to calculate medal counts for each row
host[['mbefore_hosting', 'mhosting', 'mafter_hosting']] = host.apply(calculate_medals, axis=1)

# Step 4: Create the final DataFrame 'host' with required columns
host = host[['medal_count','year_type','games', 'country_code', 'mbefore_hosting', 'mhosting', 'mafter_hosting', 'events', 'cost_per_event_mio_usd', 'athletes', 'cost_per_athlete_mio_usd', 'cost_billion_usd']]

# Step 5: reorder columns and index 
host = host.sort_values(by='year_type').reset_index(drop=True)

# Display the resulting DataFrame 'host'
host
```

```python
# Step 6: verify results in dfmc function

def verify_medal_count(dfmc, country_code, year_type):
    medal_count_check = dfmc[(dfmc['country_code'] == country_code) & (dfmc['year_type'] == year_type)]['medal_count'].values
    if len(medal_count_check) > 0:
        print(f"Medal count for {country_code} in {year_type}: {medal_count_check[0]}")
    else:
        print(f"No medal count available for {country_code} in {year_type}")

```

```python
# remove irrelvant colomns
host = host.drop(columns=['medal_count'])
host.head()
```

```python
# Figuring out, is there an overall advantage to hosting the games, what if any is the impact on the medal winnings chances. 

# Calculate pre-hosting advantage
host['pre_hosting_adv'] = ((host['mhosting'] - host['mbefore_hosting']) / host['mbefore_hosting']) * 100

# Calculate post-hosting advantage
host['post_hosting_adv'] = ((host['mafter_hosting'] - host['mhosting']) / host['mhosting']) * 100

# Reorder columns
host = host[['year_type', 'country_code','games', 'mbefore_hosting', 'mhosting', 'mafter_hosting', 'pre_hosting_adv', 'post_hosting_adv', 'events', 'cost_per_event_mio_usd', 'athletes', 'cost_per_athlete_mio_usd', 'cost_billion_usd']]

# Display the resulting DataFrame 'host'
host
```

```python
#Hypo1) compare mean from before to hosting =  H0 means are lessthan/equal. i.e. there is NO home advantage
# H1 mean hosting>mean before

#Hypo2) compare mean from after to hosting =  H0 means are lessthan/equal. i.e. there is NO home advantage
# H1 mean hosting<mean after

#Hypo3) Compare the mean from before to the mean of hosting. H0 means are lessthan/equal. i.e. there is home advantage
# H1 mean hosting>mean before
```

```python
from scipy.stats import ttest_1samp
```

```python
# Hypo1
stat, pval = ttest_1samp(host.mbefore_hosting.dropna(), 94.4)
print('stat is for the one-tailed test is ', stat)
print('pvalue for the one-tailed test is ', pval/2)
```

```python
# pvalue is greater than alpha=0.05 , cannot reject H0. >> Hypo 1 is rejected! there is NO home advantage compared to the games before
```

```python
# Hypo 2
stat, pval = ttest_1samp(host.mafter_hosting.dropna(), 94.4)
print('stat is for the one-tailed test is ', stat)
print('pvalue for the one-tailed test is ', pval/2)
```

```python
# pvalue is greater than alpha=0.05 , cannot reject H0. >> Hypo 2 is rejected! there is NO home advantage compared to the games before
```

```python
# Hypo 3
stat, pval = ttest_1samp(host.mhosting.dropna(), 59.64)
print('stat is for the one-tailed test is ', stat)
print('pvalue for the one-tailed test is ', pval/2)
```

```python
# pvalue is greater than alpha=0.05 , cannot reject H0. >> Hypo 3 is rejected! there is NO home advantage compared to the games before
```
