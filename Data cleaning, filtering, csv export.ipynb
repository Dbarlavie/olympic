```python
df4=pd.read_csv("original athlete_events-AHW.csv")
df4 = pd.DataFrame(df4)
df4.shape
```

```python
df4.columns = [column.lower().replace(' ', '_').replace('.', '').replace(',', '') for column in df4.columns]
df4.head(100)
```

```python
df4 = df4.sort_values(by='year')  
df4.head(2)
```

```python
df4.reset_index(drop=True, inplace=True)
df4.index += 0
df4.head(2)
```

```python
#Check NaN's
df4.isna().sum()
```

```python
df4['sport'].value_counts()
```

```python
# drop rows of sports which apear very little times - Military Ski Patrol, Cricket, Croquet, Motorboating, Racquets, Jeu De Paume, Roque, Basque Pelota, Aeronautics
# drop irrelevant columns

sports_to_drop = ['Military Ski Patrol', 'Cricket', 'Croquet','Motorboating','Racquets', 'Jeu De Paume','Roque','Basque Pelota','Aeronautics']
df4 = df4.drop(df4[df4['sport'].isin(sports_to_drop)].index)
df4.shape
```

```python
df4.drop(columns=['city','event','name','team','id','games'], inplace=True)
df4.head()
```

```python
df4.shape
```

```python
# Check if nan values are sport specific, country specific
# fill in those nans by gender mean

filtered_df = df4[df4[['age', 'height', 'weight']].isnull().any(axis=1)]
filtered_df.head(10)
```

```python
pivot_table = pd.pivot_table(filtered_df, index='sport', values=['age', 'height', 'weight'])
nan_counts = {}

for column in pivot_table.columns:
    nan_counts[column] = pivot_table[column].isnull().sum()

nan_counts_per_sport = filtered_df.groupby('sport')[['age', 'height', 'weight']].apply(lambda x: x.isnull().sum())

```

```python
# The biggest amount of NaNs are in the Gymnastic sport
# fill the Nan values according to the sex average
df4['age'] = df4.groupby('sex')['age'].transform(lambda x: x.fillna(x.mean()))
df4['height'] = df4.groupby('sex')['height'].transform(lambda x: x.fillna(x.mean()))
df4['weight'] = df4.groupby('sex')['weight'].transform(lambda x: x.fillna(x.mean()))
```

```python
df4.isna().sum()
```

```python
# encode medals to gold = 1, silver = 2, bronze=3
medal_mapping = {'Gold': 1, 'Silver': 2, 'Bronze': 3}
df4['medal'] = df4['medal'].map(medal_mapping)
```

```python
# fill na values within Medals to 0
df4['medal'] = df4['medal'].fillna(0)
df4.isna().sum()
```

```python
df4.medal.value_counts()
```

```python
pivot_tableM = df4.pivot_table(index='noc', columns='medal', aggfunc='size', fill_value=0).reset_index()

pivot_tableM.rename(columns={'noc': 'country_code'}, inplace=True)

# Rename columns to represent medal types as 0, 1, 2, 3
pivot_tableM.columns = ['country_code', 'Medal Type - 0', 'Medal Type - 1', 'Medal Type - 2', 'Medal Type - 3']

pivot_tableM.head()
```

```python

# Export to csv
pivot_tableM.to_csv("df4medals.csv", index=False)
```
